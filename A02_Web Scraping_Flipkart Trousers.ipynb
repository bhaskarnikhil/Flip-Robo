```python
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException
from selenium.webdriver.support import expected_conditions as EC
import os
import shutil
import requests
```

```python
from webdriver_manager.chrome import ChromeDriverManager
```

```python
driver=webdriver.Chrome(ChromeDriverManager().install())
```

```python
my_page=driver.get("https://www.flipkart.com/clothing-and-accessories/bottomwear/trouser/men-trouser/pr?sid=clo%2Cvua%2Cmle%2Clhk&otracker=categorytree&p%5B%5D=facets.occasion%255B%255D%3DFormal&otracker=nmenu_sub_Men_0_Formal%20Trousers")
```

```python
def make_directory(dirname):
    current_path=os.getcwd()
    path=os.path.join(current_path,dirname)
    if not os.path.exists(path):
        os.makedirs(path)
```

```python
make_directory('flipkart_trousers')
```

```python
urls=[]
for i in range (0,3):
    images=driver.find_elements_by_xpath("//img[@class='_3togXc']")
    try:
        for image in images:
            source=image.get_attribute('src')
            urls.append(source)
            print(source)
            if i<2:
                if (len(urls)%40)==0:
                    button=driver.find_elements_by_xpath("//a[@class='_3fVaIS']")[i].click()
                    currentURL=driver.current_url
    except StaleElementReferenceException as e:
        pass
print("Total images available for downloading is",len(urls))
```

```python
for index,link in enumerate(urls):
    print("Downloading {0} of {1} images".format(index+1,len(urls)))
    response=requests.get(link)
    with open('flipkart_trousers/img_{0}.jpeg'.format(index),"wb") as file:
        file.write(response.content)
driver.quit()
```
